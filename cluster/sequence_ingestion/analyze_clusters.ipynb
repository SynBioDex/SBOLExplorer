```python
def parse_clusters(s):
    lines = s.splitlines()
    clusters = {}
    
    for line in lines:
        part, cluster = list(map(int, line.split()))
        if cluster not in clusters:
            clusters[cluster] = set()
        clusters[cluster].add(part)
    
    return clusters


def print_clusters_stats(clusters):
    print('# of clusters: ' + str(len(clusters)))
    
    sizes = []
    
    for cluster in clusters:
        sizes.append(len(clusters[cluster]))
    
    total = 0
    for i in sizes:
        total += i
    print('total parts: ' + str(total))
    
    sizes.sort()
    print(sizes)
    
```

```python
f = open('clusters_p1.txt', 'r')
clusters = parse_clusters(f.read())
print_clusters_stats(clusters)
```
